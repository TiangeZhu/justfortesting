# 3. View selected tune families split into different clusters, and view prominent patterns by each cluster of songs

## 3.1 Check a selection of tune family, in which tunes are clustered

```python
dfcluster = pd.read_csv('selected_tf_clustered.csv')
dfcluster
```

### Take 'Wat_zag_ik_daar_van_verre_1' as tune family example.

### You may change the name in the first line of the next block with one name from the following to view results for another tune family.

```python
selected_tf = 'Wat_zag_ik_daar_van_verre_1'

df_thistf = dfcluster.loc[dfcluster['tunefamily'] == selected_tf]
df_thistf
```

```python
print("For tune family: ", selected_tf,"\n")
for cluster_id in range(3):
    dftemp = df_thistf.loc[df_thistf['cluster_num'] == cluster_id]
    print("These tunes:", dftemp["songids"].to_list(), "are in the cluster", cluster_id)
    list_salient_patterns = dftemp["salientpat"].to_list()
    print("and the followings are the most salient patterns of cluster", cluster_id,":")
    for item in list_salient_patterns:
      print(item)
    print('\n')
```
